```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('DailyDelhiClimateTrain.csv', parse_dates=['date'])
df = df.sort_values('date')

# Plot each time series
plt.figure(figsize=(16, 10))

variables = ['meantemp', 'humidity', 'wind_speed', 'meanpressure']

for i, var in enumerate(variables):
    plt.subplot(2, 2, i + 1)
    plt.plot(df['date'], df[var], label=var)
    plt.title(f'{var} over Time')
    plt.xlabel('Date')
    plt.ylabel(var)
    plt.grid(True)
    plt.tight_layout()

plt.suptitle('Initial Time Series Visualization', fontsize=16, y=1.02)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("DailyDelhiClimateTrain.csv", parse_dates=["date"])
df.set_index("date", inplace=True)

# Set a window size for smoothing (e.g., 7 days for weekly average)
window_size = 7

# Create subplots for each variable
fig, axes = plt.subplots(4, 1, figsize=(14, 12), sharex=True)
features = ["meantemp", "humidity", "wind_speed", "meanpressure"]
titles = ["Mean Temperature", "Humidity", "Windspeed", "Mean Pressure"]

for i, feature in enumerate(features):
    ax = axes[i]
    ax.plot(df.index, df[feature], label="Original", color="gray")
    ax.plot(df.index, df[feature].rolling(window=window_size, center=True).mean(),
            label=f"{window_size}-day Rolling Avg", color="blue")
    ax.set_title(titles[i])
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot histograms for each weather feature
features = ["meantemp", "humidity", "wind_speed", "meanpressure"]
titles = ["Mean Temperature", "Humidity", "Windspeed", "Mean Pressure"]

plt.figure(figsize=(14, 10))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    plt.hist(df[feature].dropna(), bins=30, color="skyblue", edgecolor="black")
    plt.title(f"Histogram of {titles[i]}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.grid(True)

plt.tight_layout()
plt.show()
```
